```python
# Mengimpor library yang diperlukan
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Meload dan Mengeksplor data
data = pd.read_csv("groceries.csv")
data.head()
```

```python
# Mengeksplor kolom data
data.columns
```

```python
# Membersihkan Data
# Menghapus ruang ekstra dalam namabarang
data['namabarang'] = data['namabarang'].str.strip()
```

```python
# Transaksi yang dilakukan pada dataset
basket = (data.groupby(['transaction_id', 'namabarang'])['transaction_id']
          .count().unstack().reset_index().fillna(0)
          .set_index('transaction_id'))
basket
```

```python
# Membangun model dan menganalisis hasilnya

# Membangun model
frq_items = apriori(basket, min_support = 0.05, use_colnames = True)

# Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
#print(rules.head())
rules.head(5)
```

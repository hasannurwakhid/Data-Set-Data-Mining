```python
#import Library 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#mengimport dataset 
dataSet = pd.read_csv("score.csv")
dataSet.keys()
```

```python
dataset = pd.DataFrame(dataSet)
dataset.head()
```

```python
#konversi ke data Array
X = np.asarray(dataSet)
print(X)
```

```python
#menampilkan data dalam bentuk scatter plot
plt.scatter(X[:,0], X[:,1], label='True Position')
plt.xlabel("Hours")
plt.ylabel ("Scores")
plt.title("Grafik Penyebaran Jumlah Jam Belajar Pelajar")
plt.show()
```

```python
#mengaktifkan K-Means dengan jumlah k=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
#plot Data point, memvisualisasikan hasil klasterisasi Hours dan Scores
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Grafik Hasil Klasifikasi Data Gaji dan Pengeluaran Konsumen")
plt.show()
```

```python
#plot Data point, memvisualisasikan hasil klasterisasi dengan centroid dari masing" klaster
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Grafik Hasil Klasifikasi Data Gaji dan Pengeluaran Konsumen")
plt.show()
```

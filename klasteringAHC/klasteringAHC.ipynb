```python
# import library
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
# Membaca dataset
data = pd.read_csv("Mall_Customers.csv")
```

```python
# Menampilkan 5 data teratas
data.head()
```

```python
# Drop kolom CustomerID dan Gender
data = data.drop(['CustomerID', 'Gender'], axis=1)
```

```python
# Menampilkan 5 data teratas
data.head()
```

```python
# buat array 2D ke dalam variabel x. Data ini yang akan kita klaster.
x = np.array(data)
```

```python
# Menampilkan variabel x yang telah berisi array numpy
print(x)
```

```python
# lakukan hierachical klastering
Z = linkage(x)
```

```python
# plot dendogram
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
